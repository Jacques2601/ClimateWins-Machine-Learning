```python
#This computes the loss function for the gradiant descent. DO NOT CHANGE!

def compute_cost(X, y, theta=np.array([[0],[0]])):
    """Given covariate matrix X, the prediction results y and coefficients theta
    compute the loss"""
    
    m = len(y)
    J=0 # initialize loss to zero
    
    # reshape theta
    theta=theta.reshape(2,1)
    
    # calculate the hypothesis - y_hat
    h_x = np.dot(X,theta)
    #print(h_x)
    
    # subtract y from y_hat, square and sum
    error_term = sum((h_x - y)**2)
    
    # divide by twice the number of samples - standard practice.
    loss = error_term/(2*m)
    
    return loss
```

```python
#This is the gradiant descent function. DO NOT CHANGE!

def gradient_descent(X, y, theta=np.array([[0],[0]]),
                    alpha=0.01, num_iterations=1500):
    """
    Solve for theta using Gradient Descent optimiztion technique. 
    Alpha is the learning rate
    """
    m = len(y)
    J_history = []
    theta0_history = []
    theta1_history = []
    theta = theta.reshape(2,1)
    
    for i in range(num_iterations):
        error = (np.dot(X, theta) - y)
        
        term0 = (alpha/m) * sum(error* X[:,0].reshape(m,1))
        term1 = (alpha/m) * sum(error* X[:,1].reshape(m,1))
        
        # update theta
        term_vector = np.array([[term0],[term1]])
        #print(term_vector)
        theta = theta - term_vector.reshape(2,1)
        
        # store history values
        theta0_history.append(theta[0].tolist()[0])
        theta1_history.append(theta[1].tolist()[0])
        J_history.append(compute_cost(X,y,theta).tolist()[0])
        
    return (theta, J_history, theta0_history, theta1_history)
```

```python
%%time
#This runs your data through a gradiant descent for the starting conditions in 'theta_init.'
#You will need to adjust these numbers

num_iterations=1000 #<---Decide how many iterations you need. Start small and work up. Over 10,000 iterations will take a few seconds.
theta_init=np.array([[-10],[-5]]) #<---this is where you put the guess for [theta0], [theta1]. Start with 1 and 1.
alpha=0.01 #<---Decide what your step size is. Try values between 0.1 and 0.00001. You will need to adjust your iterations.
#If your solution is not converging, try a smaller step size.
theta, J_history, theta0_history, theta1_history = gradient_descent(X,y, theta_init,
                                                                   alpha, num_iterations)
```

```python
def summarize_run(X, y, theta, J_history, alpha, num_iterations, name=""):
    m = len(y)
    y_hat = X @ theta
    residuals = y_hat - y

    # Core metrics
    final_loss = float(J_history[-1])                 # J(θ) at the end
    rmse = float(np.sqrt((residuals**2).mean()))      # Root-mean-square error
    mae = float(np.abs(residuals).mean())             # Mean absolute error

    # Convergence diagnostics
    last_k = min(20, len(J_history)-1)
    dJ = float(J_history[-1] - J_history[-1-last_k]) if last_k > 0 else np.nan
    grad = np.array([
        (residuals * X[:,0:1]).mean(),  # dJ/dθ0 (up to scaling)
        (residuals * X[:,1:2]).mean()   # dJ/dθ1 (up to scaling)
    ]).reshape(2,)
    grad_norm = float(np.linalg.norm(grad))

    # Simple R^2 (for a quick sanity check)
    ss_res = float(((y - y_hat)**2).sum())
    ss_tot = float(((y - y.mean())**2).sum())
    r2 = 1 - ss_res/ss_tot if ss_tot > 0 else np.nan

    # Slope sanity: predicted y at beginning vs end of year-index
    x_min = X[:,1].min(); x_max = X[:,1].max()
    y_min = float(theta[0,0] + theta[1,0]*x_min)
    y_max = float(theta[0,0] + theta[1,0]*x_max)

    print(f"=== {name or 'Run'} ===")
    print(f"alpha={alpha}, iterations={num_iterations}")
    print(f"theta0={theta[0,0]:.6f}, theta1={theta[1,0]:.6f}")
    print(f"final_loss={final_loss:.6f} | RMSE={rmse:.6f} | MAE={mae:.6f} | R^2={r2:.4f}")
    print(f"ΔJ over last {last_k} iters = {dJ:.6e}  |  ||grad||≈{grad_norm:.6e}")
    print(f"Predicted y range across year-index: [{y_min:.3f}, {y_max:.3f}]")
    return {
        "alpha": alpha, "iterations": num_iterations,
        "theta0": float(theta[0,0]), "theta1": float(theta[1,0]),
        "final_loss": final_loss, "rmse": rmse, "mae": mae, "r2": r2,
        "delta_J_last_k": dJ, "grad_norm": grad_norm,
        "y_hat_min": y_min, "y_hat_max": y_max
    }
```
